```
class Activation_Sigmoid:
  def forward(self,inputs):
    self.output=1/(1 + np.exp(-inputs))
  def gradient(self,outp):
    self.deltaoutput=outp * (1 - outp)
  def printActivationLayerProperties(self):
    print("Activation Layer Properties")
    print("h",self.output.shape)
    print("deltah",self.deltaoutput.shape)
```

```
class CrossEntropy():
    def __init__(self, y_true, y_pred):
        self.y_true = y_true
        self.y_pred = y_pred
    def calculateLoss(self):
        return -np.sum(self.y_true * np.log(self.y_pred))

    def grad(self):
        return -(self.y_true/(self.y_pred + 1e-9))
```

```
class Activation_Softmax:
  def forward(self,inputs):
    numerator_values=np.exp(inputs-np.max(inputs))
    probabilites=numerator_values/np.sum(numerator_values)
    self.output=probabilites
```

```
import numpy as np
class Layer:
  def __init__(self,n_inputs,n_neurons):
    self.weights=0.10*np.random.randn(n_inputs,n_neurons)
    self.biases=np.zeros((1,n_neurons))
    self.deltaWeights= np.zeros((n_inputs,n_neurons))
    self.deltaBiases= np.zeros((1,n_neurons))
    self.deltaOutput=np.zeros((1,n_neurons))
  def getWeights(self):
    return self.weights
  def getBiases(self):
    return self.biases
  def forward(self,inputs):
    self.output= np.dot(inputs,self.weights)+self.biases
  def delta(self,delWeights,delbiases,deloutput):
    self.deltaWeights= delWeights
    self.deltaBiases= delbiases
    self.deltaOutput = deloutput
  def update_StochsticGradientDescent(self,eta):
    print("updateWeights")
    self.weights= self.weights- eta*self.deltaWeights
    self.biases=self.biases-eta*self.deltaBiases
  def printLayerProperties(self):
    print("weights ",self.weights.shape)
    print("biases ",self.biases.shape)
    print("ouput-a",self.output.shape)
    print("deltaWeights",self.deltaWeights.shape)
    print("deltaBiases",self.deltaBiases.shape)
    print("delta-a",self.deltaOutput.shape)
```
