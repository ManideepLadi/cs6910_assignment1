```
class Activation_Sigmoid:
  def forward(self,inputs):
    self.output=1/(1 + np.exp(-inputs))
```

```
#Testing sigmoid activation fuction
import numpy as np
a=np.random.randn(10)
sigmoid=Activation_Sigmoid();
sigmoid.forward(a)
print(sigmoid.output)
```

```
class Activation_Softmax:
  def forward(self,inputs):
    numerator_values=np.exp(inputs-np.max(inputs))
    print(np.sum(numerator_values))
    probabilites=numerator_values/np.sum(numerator_values)
    self.output=probabilites
```

```
import numpy as np
a=np.random.randn(10)
softmax=Activation_Softmax();
softmax.forward(a)
print(sum(softmax.output))
```

```
import numpy as np
class Layer:
  def __init__(self,n_inputs,n_neurons):
    self.weights=0.10*np.random.randn(n_inputs,n_neurons)
    self.biases=np.zeros((1,n_neurons))
  def forward(self,inputs):
    self.output= np.dot(inputs,self.weights)+self.biases
  def update(self,eta,deltaWeights,deltabiases):
    self.weights= self.weights- eta*deltaWeights
    self.biases=self.biases-eta*deltabiases
```

```
def train_model(NoOfLayers):
  #feed forward
  HiddenLayers=[]
  HiddenLayersActivation=[]
  for i in range(NoOfLayers-1):
    HiddenLayers.append(Layer(784,784))
    HiddenLayersActivation.append(Activation_Sigmoid())
  
  HiddenLayers[0].forward(X_train[1])
  HiddenLayersActivation[0].forward(HiddenLayers[0].output)
  print("FirstHiddenLayer:")
  print(HiddenLayersActivation[0].output.shape)
  for j in range(NoOfLayers-2):
    print("HiddenLayer",j+2)
    HiddenLayers[j+1].forward(HiddenLayersActivation[j].output)
    print(HiddenLayers[j+1].output.shape)
    HiddenLayersActivation[j+1].forward(HiddenLayers[j+1].output)
    print(HiddenLayersActivation[j+1].output.shape)
  print("OutputLayer")
  opLayer=Layer(784,10)
  opLayer.forward(HiddenLayersActivation[NoOfLayers-2].output)
  print(opLayer.output.shape)
  opActivation=Activation_Softmax()
  opActivation.forward(opLayer.output)
  print(sum(opActivation.output))
  print(np.argmax(opActivation.output,axis=1))
  print(y_train[1])
```
