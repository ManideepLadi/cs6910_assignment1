```
from keras.datasets import fashion_mnist

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
labels=["T-shirt/top","Trouser","Pullover","Dress","Coat",
        "Sandal","Shirt","Sneaker","Bag","Ankle boot"]
```

```
def plot_images():
  plottedClass =  [False for i in range(10)] 
  images=[]
  classes=[]
  for i in range(100):
      if  plottedClass[y_train[i]]==False:
        images.append(X_train[i])
        classes.append(labels[y_train[i]])
        plottedClass[y_train[i]]=True
  wandb.log({"example": [wandb.Image(img,caption=label) for img,label in zip(images,classes)]})
```

```
import wandb
wandb.init(project="fashon_mnist", entity="manideepladi")
# Log metrics with wandb
 
plot_images()



# Save model to wandb
```
